```python
beam_splits = 0
for row in range(2, len(grid)):
    line = grid[row]
    # line, row
    for col, char in enumerate(line):
        if char == "^" and grid[row - 1][col] == "|":
            beam_splits += 1
            grid[row][col - 1] = grid[row][col + 1] = "|"
        elif grid[row - 1][col] == "|":
            grid[row][col] = "|"

    # print("\n".join(["".join(line) for line in grid]))

beam_splits
```

```python
from functools import lru_cache


@lru_cache(maxsize=1000)
def count_paths(current_pos):
    """Count all unique timeline paths from current position."""
    next_pos = (current_pos[0] + 1, current_pos[1])

    # If out of bounds, this is one complete timeline
    if next_pos[0] >= len(grid):
        return 1

    if grid[next_pos[0]][next_pos[1]] == "^":
        # Quantum split: count paths from BOTH directions
        left_count = count_paths((next_pos[0], next_pos[1] - 1))
        right_count = count_paths((next_pos[0], next_pos[1] + 1))
        return left_count + right_count
    else:
        # Continue straight down
        return count_paths(next_pos)


count_paths((0, start_col))

count_paths.cache_info()
```
