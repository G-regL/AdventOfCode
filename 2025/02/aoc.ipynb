```python
#from icecream import ic

def in_grid(row, col, limits) -> bool:
    return 0 <= row < limits[0] and 0 <= col < limits[1]

productIDs = [r.split("-") for r in open('input.txt').read().split(',')]

#productIDs = [["11","22"], ["95","115"]]
#productIDs
```

```python
invalid = []
for start,end in productIDs:
    print(f"Range {start}-{end}")
    for pid in range(int(start), int(end)+1):
        print(f"  {pid}")
        pid = str(pid)
        if len(pid) % 2 != 0:
            continue
        half = int(len(pid) / 2)
        print(f"    {pid[:half]} <> {pid[half:]}")
        if pid[:half] == pid[half:]:
            invalid.append(pid)

print(invalid)
print(sum(list(map(int,invalid))))
        
```

```python
invalid = set()
for start,end in productIDs:
    #print(f"Range {start}-{end}")
    for pid in range(int(start), int(end)+1):
        #print(f"  {pid}")
        pid = str(pid)

        #print(pid[0:1])
        for chars in range(1, int(len(pid) / 2) + 1):
            #print(f"    chars {pid[0:chars]}")
            pattern_count = pid.count(pid[0:chars])
            #print(f"    pattern_count {pattern_count}")
            if pattern_count * chars == len(pid):
                #print("    whole pid is pattern?")
                invalid.add(pid)


print(invalid)
print(sum(list(map(int,invalid))))
```
