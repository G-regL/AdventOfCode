```python
lines = open("test.txt").read().split("\n")
col_widths = [len(col) - 1 for col in re.findall(r"([\+\*] +)", lines[-1])]
col_widths[-1] += 1
lines

data = []
for line in lines:
    col = 0
    row = []
    for i in col_widths:
        number = line[col : col + i]
        row.append(number)
        col += i + 1
    data.append(row)

sheet = list(zip(*data[::-1]))
sheet


total = 0
final_sheet = []
for line in sheet:
    numbers = [line[0].strip()]
    for idx in range(len(line[0])):
        # idx
        n = [char[idx] for char in line[1:]]
        n.reverse()
        numbers.append(int("".join(n)))
    final_sheet.append(numbers)
final_sheet
total = 0

for line in final_sheet:
    if line[0] == "*":
        total += prod(line[1:])
    elif line[0] == "+":
        total += sum(line[1:])

total
```
