```python
# from icecream import ic
import re

from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"


def in_grid(row, col, limits) -> bool:
    return 0 <= row < limits[0] and 0 <= col < limits[1]


data = [re.findall(r"\S+", line) for line in open("test.txt").read().split("\n")]

sheet = []
for line in data[:-1]:
    sheet.append(list(map(int, line)))

sheet.append(data[-1])
sheet = list(zip(*sheet[::-1]))
sheet
```

```python
from math import prod

total = 0
for line in sheet:
    if line[0] == "*":
        total += prod(line[1:])
    elif line[0] == "+":
        total += sum(line[1:])

total
```
