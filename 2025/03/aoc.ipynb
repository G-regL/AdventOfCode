```python
#from icecream import ic

def in_grid(row, col, limits) -> bool:
    return 0 <= row < limits[0] and 0 <= col < limits[1]

banks = open('test.txt').read().split('\n')

#banks
```

```python
joltage = 0
for bank in banks:
    jolts = "0"
    for b in range(0, len(bank) -1):
        if bank[b]> jolts:
            jolts = bank[b]
    #print(jolts)

    jolts += "0"
    for b in range(bank.find(jolts[0]) + 1, len(bank)):
        if bank[b] > jolts[1]:
            jolts = jolts[0] + bank[b]

    #print(jolts)
    joltage += int(jolts)

print(joltage)
```

```python
totaljoltage = 0
batteries_required = 12

for idx, bank in enumerate(banks):
    
    joltage = ""
    bank_len = len(bank)
    print(f"\n[Bank {idx}] raw='{bank}' (len={bank_len})")

    iteration = 0
    
    last_battery = 0
    for batteries_left in range(batteries_required, 0, -1):
        iteration += 1
        search_end = max(bank_len - batteries_left, bank_len)
        print(f"  Iter {iteration}: batteries_left={batteries_left}, searching indices {last_battery}..{search_end-1}")

        jolts = "0"

        for b in range(last_battery, search_end):
            val = bank[b]
            if val > jolts:
                jolts = val
                last_battery = b + 1
                print(f"    new best at index {b}: '{jolts}'")

        joltage += jolts
        print(f"    Selected '{jolts}' (index {last_battery}), joltage so far: {joltage}")

        batteries_left -= 1

    totaljoltage += int(joltage)
    print(f"[Bank {idx}] final selected joltage list: {joltage}")

print(f"\nTotal joltage from all banks: {totaljoltage}")
```
