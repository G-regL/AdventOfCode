### Part 1

```python
fresh_ingredients = set()
for a in available:
    for s, e in fresh_ranges:
        if s <= a <= e:
            fresh_ingredients.add(a)

len(fresh_ingredients)
```

### Part 2, but... broken

```python
fresh_ranges = sorted(fresh_ranges, key=lambda x: x[0])
known_ranges = [fresh_ranges[0]]
for fs, fe in fresh_ranges[1:]:
    print(f"\nProcessing range {fs:,}-{fe:,}")
    print(f" Known ranges {known_ranges}")
    overlap_found = False
    for idx, (ks, ke) in enumerate(known_ranges):
        print(f"  Checking overlap of known range {ks:,}-{ke:,}, {idx=}")

        if ks < fs < ke and ks < fe < ke:
            print("   Fully encompassed by known range")
            overlap_found = True
            break

        overlap_e = (
            fs <= ks and fe <= ke and fe >= ks
        )  # end is within known range, but start isn't
        overlap_s = (
            fe >= ke and fs >= ks and fs <= ke
        )  # start is with known range, but end isn't
        print(f"   {overlap_s=} {overlap_e=}")

        new_range = [ks, ke]

        if overlap_s:
            print(f"   smaller start; {fs:,} < {ks:,}; {ks - fs:,}Δ")
            new_range[0] = fs

        if overlap_e:
            print(f"   bigger end; {fe:,} > {ke:,}; {fe - ke:,}Δ")
            new_range[1] = fe

        if new_range != [ks, ke]:
            overlap_found = True
            print(f"   extended range, replacing {[ks, ke]} => {new_range}")
            known_ranges[idx] = new_range
            break

    if not overlap_found:
        print(f"  No overlap found adding {[fs, fe]} to list of known ranges")
        known_ranges.append([fs, fe])

known_ranges
fresh_ingredients = sum([(e + 1) - s for s, e in known_ranges])
fresh_ingredients
# known_ranges[0][1] - known_ranges[0][0]

# 372934381425813
# ??372934381425695
```

### Part 2, but works!

```python
fresh_ranges = sorted(fresh_ranges, key=lambda x: x[0])
known_ranges = [fresh_ranges[0]]
for fs, fe in fresh_ranges[1:]:
    if fs > known_ranges[-1][1]:
        known_ranges.append([fs, fe])
    else:
        # otherwise, extend the last range
        known_ranges[-1][1] = max(known_ranges[-1][1], fe)

sum([e - s + 1 for s, e in known_ranges])
len(known_ranges)
```
