```python
#from icecream import ic

def in_grid(row, col, limits) -> bool:
    return 0 <= row < limits[0] and 0 <= col < limits[1]

moves = [[l[0],int(l[1:]) % 100, int(l[1:])] for l in open('input.txt').read().split('\n')]

#moves
```

```python
dial_position = 50
zero_counts = 0

for m in moves:
    if m[0] == "R":
        dial_position = dial_position + m[1]
    else:
        dial_position = dial_position - m[1]
    
    dial_position = dial_position % 100
    
    if dial_position == 0:
        zero_counts += 1

print (f"Part 1: {zero_counts}")
```

```python
dial_position = 50
zero_counts_p2 = 0

for m in moves:
    dial_position_old = dial_position
    #print (f"- {m}")
    #print (f"  start dial {dial_position:3}")
    if m[0] == "R":
        dial_position = dial_position + m[1]
    else:
        dial_position = dial_position - m[1]

    dial_position = dial_position % 100
    
    if dial_position != 0 and dial_position_old != 0:
        if m[0] == "R" and dial_position_old + m[1] > 99:
            #print(f"  Crossed 0 ↷ {dial_position_old + m[1]}")
            zero_counts_p2 += 1
        elif m[0] == "L" and dial_position_old - m[1] < 0:
            #print(f"  Crossed 0 ↶ {dial_position_old - m[1]}")
            zero_counts_p2 += 1
    
    if dial_position == 0:
        #print (f"  dial at 0, incrementing")
        zero_counts_p2 += 1

    if m[2] > 100:
        rotations = int(m[2] / 100)
        #print (f"  add rotations, {rotations}")
        zero_counts_p2 += rotations 

    #print (f"  end dial {dial_position:3}; {zero_counts_p2}")
    #print(f"  dial: {dial_position}; {zero_counts=}; {zero_counts_p2=}")


print (f"Part 2: {zero_counts_p2}")

```
