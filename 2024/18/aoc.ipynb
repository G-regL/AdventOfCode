```python
from icecream import ic
from collections import deque


bytes = [(int(line.split(",")[1]), int(line.split(",")[0])) for line in open('input.txt').read().split('\n') ]
grid_size = 70

#bytes = [(int(line.split(",")[1]), int(line.split(",")[0])) for line in open('test.txt').read().split('\n') ]
#grid_size = 6

#for num_bytes in range(len(bytes) + 1):

def bfs(bytes, grid_size):
    seen = set(bytes)

    queue = deque([((0,0), set([(0,0)]))])
    shortest = float("inf")
    shortest_path = []
    while queue:
        (row, column), path = queue.popleft()

        if (row, column) == (grid_size, grid_size):
            #print(f"Found end")
            if len(path) < shortest:
                #print(f"({len(path)}) smaller than {shortest=}")
                shortest_path = path
                shortest = len(path)

        for (dr, dc) in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            row_new, column_new = row + dr, column + dc
            if 0 > row_new or row_new > grid_size or 0 > column_new or column_new > grid_size:
                #print(f"dir ({row_new},{column_new}) out of bounds")
                continue
            
            if (row_new, column_new) in bytes or (row_new, column_new) in seen:
                continue

            seen.add((row_new, column_new))

            if (row_new, column_new) not in path:
                new_path = set(path)
                #print(f"({row_new},{column_new}) not seen, and not in path")
                new_path.add((row_new, column_new))
                queue.append(((row_new, column_new), new_path))
    
        #ic(queue)
        #print("\n")
    return shortest -1, shortest_path


def draw_grid(size, bytes, path):
    for r in range(size +1):
        line = ""
        for c in range(size +1):
            if (r, c) in bytes:
                line += "#"
            elif (r, c) in path:
                line += "O"
            else:
                line += "."
        print(line)
#print(scores)

foo = 21

#print(bfs(bytes[:foo], 6, foo))

path_length, _ = bfs(bytes[:1024], grid_size)
print(f"Part 1: {path_length}")

# print(bfs(bytes[:1025], grid_size))


# Repeat until the pointers low and high meet each other
high = len(bytes) - 1
low = 1024
while low < high:
    mid = (low + high) //2

    path_length, _ = bfs(bytes[:mid], grid_size)

    if path_length == float("inf"):
        high = mid
    else:
        low = mid + 1


last = bytes[low -1]
print(f"Last byte ({last[1]},{last[0]}) caused an issue")

```
