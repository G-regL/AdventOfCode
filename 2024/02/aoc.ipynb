### Setup imports and get raw data

```python
#from icecream import ic

data = open('input.txt').read().split('\n')

def safe(levels) -> bool:
    decreasing = [levels[i-1] > v for i,v in enumerate(levels[1:], 1)]
    increasing = [levels[i-1] < v for i,v in enumerate(levels[1:], 1)]
    gaps = [0 < abs(levels[i-1] - v) < 4 for i,v in enumerate(levels[1:], 1)]

    #print(f"{levels=}")
    if (all(decreasing) or all(increasing)) and all(gaps):
        return True
    else:
        return False

safe_P1 = 0
safe_P2 = 0
for report in data:

    levels = list(map(int, report.split()))
    if safe(levels):
        print("Safe P1")
        safe_P1 += 1


    for r in range (len(report)):
        # I slice this way. I slice that way. I take out middle. Goodbye stupid useless number. haha!
        new_levels = levels[:r] + levels[r + 1 :]

        print(f"{levels=}, {new_levels=}")
        if safe(new_levels):
            print("Safe P2")
            safe_P2 += 1
            break




    print("")
print(safe_P1)
print(safe_P2)
```
