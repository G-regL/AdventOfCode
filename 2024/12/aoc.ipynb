```python
def dfs(grid, start_row, start_col, letter, visited):
    # Directions to move in the grid (up, down, left, right)
    directions = [(1,0), (-1,0), (0,1), (0,-1)]
    stack = [(start_row, start_col)]
    coordinates_set = set()
    
    # While the stack is not empty, keep checking adjacent coordinates
    while stack:
        row, col = stack.pop()
        if (0 <= row < len(grid)) and (0 <= col < len(grid[0])) and grid[row][col] == letter and (row, col) not in visited:
            # Mark this cell as visited
            visited.add((row, col))
            coordinates_set.add((row, col))
            
            # Add adjacent cells to the stack
            for dr, dc in directions:
                adj_row, adj_col = row + dr, col + dc
                stack.append((adj_row, adj_col))
                
    return coordinates_set


grid = open('input.txt').read().split('\n')
# Create a set to keep track of visited cells
visited = set()
# A dictionary to keep sets of coordinates with the same letter
plants_regions = {}

# Iterate through each cell in the grid to perform DFS
for row in range(len(grid)):
    for col in range(len(grid[0])):
        if (row, col) in visited:
            continue

        letter = grid[row][col]
        same_set = dfs(grid, row, col, letter, visited)
        
        if letter in plants_regions:
            plants_regions[letter].append(same_set)
        else:
            plants_regions[letter] = [same_set]
    

print(plants_regions)

total_price = 0

for plant, regions in plants_regions.items():
    for region in regions:
        
        # Directions to check neighboring cells
        directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]
        perimeter = 0
        
        # Iterate over each cell in the set
        for (row, col) in region:
            for dr, dc in directions:
                neighbor = (row + dr, col + dc)
                # Increase perimeter if the neighbor cell is not in the set (edge or boundary)
                if neighbor not in region:
                    perimeter += 1


        print(f"{plant=}, {len(region)=}, {perimeter=}, {len(region) * perimeter}")

        total_price += len(region) * perimeter

print(total_price)

```
